## Data Exploration - Understand the data - Investigate the variables

```python
#defines the different reponses in the dataset's "claim_status" column 
uniqueclaimscategories = data['claim_status'].unique()
print(uniqueclaimscategories)

#shows how many counts of each dataset of claims and displays the data type for each
data['claim_status'].value_counts()
```

**Notes:** The values shown show a even distribution of both "claim" and "opinion" categories within the claim_status column

```python
# What is the average view count of videos with "claim" status?
# What is the average view count of videos with "opinion" status?
### YOUR CODE HERE ###

statusclaimviewcount = (data['claim_status'] == "claim") | (data['claim_status'] =="opinion") #groups the claim status for both claim and opinion into statusclaimviewcount with a mask

data[statusclaimviewcount].groupby('claim_status').agg({'video_view_count': ['mean','median']}) #groups the claims by claim_status, then displays the average and median of the claims together
```

**Notes:** The claim category has significantly higher mean and median claim status's than the opinion category

```python
# Get counts for each group combination of claim status and author ban status
### YOUR CODE HERE ###
data.groupby(["claim_status",'author_ban_status']).count()
data.groupby(["claim_status",'author_ban_status']).count()["video_id"]
```

**Notes:** The author banned 

```python
# What's the median video share count of each author ban status?
### YOUR CODE HERE ###
data.groupby('author_ban_status')['video_share_count'].agg("median")
```

**Notes:** Banned features significantly more video share count than active users

```python
#creates a table showing video_view_count, like_count and share_count by "author_ban_status"
data.groupby('author_ban_status').agg({'video_view_count':  ("count","mean","median"),
                                       'video_like_count':  ("count","mean","median"),
                                       'video_share_count': ("count","mean","median")})
```

**Notes:** Banned users have significantly more views, likes and shares than active users

```python
#creates "likes_per_view" column in the dataset
data["likes_per_view"] = (data['video_like_count'] / data["video_view_count"])

#creates "comments_per_view" column in the dataset
data["comments_per_view"] = (data["video_comment_count"]) / data["video_view_count"]

#creates "shares_per_view" column in the dataset
data['shares_per_view'] = (data["video_share_count"]) / data["video_view_count"]

#prints dataframe to ensure new categories are inserted in to the dataset
data.head(3)

```

```python
#groups both claim status and author ban status from the dataframe and then aggregates the categories "likes_per_view", "comments_per_view","shares_per_view" and displays each categories: 
# Mean, Median, and Count
data.groupby(['claim_status','author_ban_status']).agg({"likes_per_view": ("count","mean","median"),
                                                        "comments_per_view": ("count","mean","median"),
                                                        "shares_per_view": ("count","mean","median")})
```

**Notes** After making new categories, we can see that "opinion" offers a higher active count, while its counts in likes, comments, and shares are signficantly less than the banned users in claim
